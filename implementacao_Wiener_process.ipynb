```python
import numpy.random as npr
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def wiener_process(dt=0.1, x0=0, n=1000):

    # W(t=0) = 0
    # initialize W(t) with zeros
    W = np.zeros(n+1)

    # we create N+1 timesteps: t=0,1,2,3...N
    t = np.linspace(x0, n, n+1)

    #we have to use cumulative sum: on every step the additional value is
    # draw from a normal distribution wwith mean 0 and variance dt ... N(0, dt)
    # btw: N(0,dt) = sqrt(dt)*N(0,1) usually this formula is used !!!
    W[1:n + 1] = np.cumsum(npr.normal(0, np.sqrt(dt), size=n)) # aqui, dt é o devio padrão

    return t, W

def plot_wiener_process(t, W):
    plt.plot(t, W)
    plt.xlabel('time (t)')
    plt.ylabel('W(t)')
    plt.title('Wiener Process')
    plt.show()


if __name__ == '__main__':
    t, W = wiener_process()
    plot_wiener_process(t, W)
```
